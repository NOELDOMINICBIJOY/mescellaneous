```python
q= 1e-10
epsilon_0 = 8.8e-12
a = 0.1
```

```python
# Import required modules
import numpy as np
import matplotlib.pyplot as plt

# 1D arrays
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)

# Meshgrid
X,Y = np.meshgrid(x,y)

# Assign vector directions
Ex = ((q*X)/(4*np.pi*epsilon_0))*((1/((X**2)+(Y-a)**2)**(3/2)) - (1/((X**2)+(Y+a)**2)**(3/2)))

Ey = ((q)/(4*np.pi*epsilon_0))*(((Y-a)/((X**2)+(Y-a)**2)**(3/2)) - ((Y+a)/((X**2)+(Y+a)**2)**(3/2)))

# Depict illustration
plt.figure(figsize=(10, 10))
plt.streamplot(X,Y,Ex,Ey, density=1.4, linewidth=None, color='#A23BEC')
plt.plot(0,.1,'-or')
plt.plot(0,-.1,'-og')
plt.title('Electromagnetic Field')

# Show plot with grid
plt.grid()
plt.show()
X
```

```python
# r>>a

x = np.arange(-100,100,0.1)
y = np.arange(-100,100,0.1)

# Meshgrid
X,Y = np.meshgrid(x,y)

Ex = (3*q*X*Y*a/2*np.pi*epsilon_0*((x**2 + y**2)**(5/2)))
Ey = (q/4*np.pi*epsilon_0*((x**2 + y**2)**(3/2))*((6*(Y**2)*a)/((x**2 + y**2))-(2*a)))


# Depict illustration
plt.figure(figsize=(10, 10))
plt.streamplot(X,Y,Ex,Ey, density=1.4, linewidth=None, color='#AAA231')


# Show plot with grid
plt.grid()
plt.show()
```

```python
k = 1/(4*np.pi*epsilon_0)
```

```python
#single charge
x = np.arange(-100,100,0.1)
y = np.arange(-100,100,0.1)

# Meshgrid
X,Y = np.meshgrid(x,y)

Ex = (k*+q*X)/(X**2 + Y**2)**(3/2)

Ey = (k*+q*Y)/(X**2 + Y**2)**(3/2)

# Depict illustration
plt.figure(figsize=(10, 10))
plt.streamplot(X,Y,Ex,Ey, density=1.4, linewidth=None, color='#034206')
plt.title('Electromagnetic Field')

# Show plot with grid
plt.grid()
plt.show()
```
