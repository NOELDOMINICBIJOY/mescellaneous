```python
import matplotlib.pyplot as plt
import numpy as np
import cmath

plt.style.use(['science', 'notebook', 'grid'])
```

```python
n = 2
z = np.arange(0,2*np.pi,0.0001)
func = []
for i in z:
    re = i - np.sin(i)
    im = 1 - np.cos(i)
    z_i = complex(re, im)
#     h_factor = complex(np.cos(2*np.pi*n*z_i), -np.sin(2*np.pi*n*z_i))
#     z_i *= h_factor
    func.append(z_i)
    
func = np.array(func)
```

```python
plt.figure(figsize=(20,6))
plt.plot(func.real,func.imag, marker='.', color='green', label='output')

plt.xlabel = f"real"
plt.ylabel = f"imaginary"
```
