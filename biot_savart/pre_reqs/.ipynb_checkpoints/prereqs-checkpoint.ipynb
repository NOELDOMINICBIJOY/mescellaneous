```python
# derivatives 

x = np.arange(0, 10, 0.01)
y = np.cos(x)

def derivative(x, f):
    delta_x = x[1] - x[0]
    derivative = []
    
    for index, element in enumerate(f):
        if index == 0:
            derivative.append(0)
        else:
            delta_f = f[index] - f[index - 1]
            derivative.append(delta_f/delta_x)
        
    return np.array(derivative)
    
```

```python
plt.plot(x, y)
plt.plot(x, derivative(x,y))
```

```python
# derivative of curves
t = np.arange(0, 1, 0.01)
x = np.cos(5*t)
y = np.sin(5*t)
z = 0.2*t

s = np.array([x, y, z])


fig = plt.figure()
ax = plt.axes(projection='3d')

plt.plot(s[0], s[1], s[2])
```

```python
def derivative_curve(t, s, r):
    x_d = derivative(t, s[0])
    y_d = derivative(t, s[1])
    z_d = derivative(t, s[2])
    
    derivative_s = np.array([x_d, y_d, z_d])
    return derivative_s

der_s = derivative_curve(t, s)
plt.plot(der_s[0], der_s[1], der_s[2])
```
