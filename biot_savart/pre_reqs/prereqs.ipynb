```python
# curves
# curves are just a collection of 3d points
# example 

import matplotlib.pyplot as plt
import scienceplots

plt.style.use(['science', 'notebook', 'dark_background'])
plt.style.use(['science', 'notebook', 'dark_background', 'grid'])
plt.style.use(['science', 'notebook', 'dark_background'])

t = np.linspace(0, 10, 1000)

def x(t):
    return 0.8*np.cos(5*t)
def y(t):
    return 0.5*np.sin(5*t)
def z(t):
    return 0*t

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(x(t), y(t), z(t), 'r')

t_0 = 0

ax.plot3D(x(t_0), y(t_0), z(t_0), 'o', color='cyan')
```

```python
def curve(x, y, z):
    curve = []
    index = 0
    
    while index < x.shape[0]:
        curve.append([x[index], y[index], z[index]])
        index += 1
    
    return np.array(curve)

curve = curve(x(t), y(t), z(t))      
```

```python
def tangent_to_curve(curve, parametarising_variable):
    delta_t = parametarising_variable[1] - parametarising_variable[0]
    tangents = []
    
    for index, element in enumerate(curve):       
        if index < curve.shape[0] - 1 :
            delta_s = curve[index + 1] - curve[index]
            tangents.append(delta_s)           
        
    return np.array(tangents)
```

```python
tangents = tangent_to_curve(curve, t)
```

```python
# vector plotting
import matplotlib.pyplot as plt
import numpy as np

soa = np.array([[0, 0, 10, 0.5, 0.7, 0]])
X, Y, Z, U, V, W = zip(*soa)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

ax.quiver(X, Y, Z, U, V, W, color='red')
ax.set_xlim([-1, 0.5])
ax.set_ylim([-1, 1.5])
ax.set_zlim([-1, 8])
plt.show()
```

```python
# tangent vector plotting

point = 962

fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')
ax.plot3D(x(t), y(t), z(t), '#00ff33')

t_0 = t[point]
origin_and_tangent = [x(t_0), y(t_0), z(t_0), 10*tangents[point][0], 10*tangents[point][1], 10*tangents[point][2]]

soa = np.array([origin_and_tangent])
X, Y, Z, U, V, W = zip(*soa)

ax.quiver(X, Y, Z, U, V, W, color='cyan')
ax.plot3D(x(t_0), y(t_0), z(t_0), 'o', color='cyan')

ax.set_xlim([-1, 0.5])
ax.set_ylim([-1, 1.5])
ax.set_zlim([-1, 8])
plt.show()
```
