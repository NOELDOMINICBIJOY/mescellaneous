```python
from scipy.integrate import quad
import numpy as np
import matplotlib.pyplot as plt
```

```python
a= 10
x_0 =1

x = np.arange(-5, 5,0.0001)

def y_1(x):
    return np.e**(1j*x)


plt.plot(x,y_1(x), color='green', label='approximation for delta function')
plt.xlabel= 'x'
plt.ylabel = 'y'
plt.legend()
plt.grid()
```

```python
def y_1(x):
    return (1/(np.sqrt((np.pi)*a)))*np.e**((-((x-x_0)**2))/(a**2))

def get_y_FT(y_1, k):
    y_FT_integrand_real = lambda x: np.real(y_1(x)*np.exp(-2*np.pi*1j*k*x))
    y_FT_integrand_comp = lambda x: np.imag(y_1(x)*np.exp(-2*np.pi*1j*k*x))
    y_FT_real = quad(y_FT_integrand_real, -np.inf, np.inf)[0]
    y_FT_comp = quad(y_FT_integrand_comp, -np.inf, np.inf)[0]
    return y_FT_real + 1j*y_FT_comp
```

```python
k = np.linspace(-4, 4, 100)
y_FT = np.vectorize(get_y_FT)(y_1, k)
```

```python
plt.plot(k, np.abs(y_FT))
plt.grid()
```
